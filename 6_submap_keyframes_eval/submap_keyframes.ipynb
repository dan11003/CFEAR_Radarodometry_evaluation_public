```python
############# LOAD DATA #########################
#### Activate with esc+r, disable with esc+y ####


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import csv
from pandas.api.types import CategoricalDtype
```

```python

df = pd.read_csv('./data/submap_eval_cfear_2022-06-02_1147_eval.csv', sep=r',', header=0)
```

```python
sns.set_theme(style="ticks", color_codes=True)
sns.set(style="ticks")
colors = ["#E59866", "#D35400", "#873600", "#85C1E9", "#3498DB", "#2874A6"]
colors_two = [ "#D35400", "#3498DB", "#2ECC71"]
pal=sns.color_palette(colors_two)

#print(df)
df = df.rename(columns={'Registration avg': 'Registration time [ms]'})
df = df.rename(columns={'Surface points avg': 'Surface points'})
df = df.rename(columns={'Trans.err.(%)': 'Transl. Error[%]'})
df = df.rename(columns={'Rot.err.(deg/100m)': 'Rot. Error[deg/100m]'})
df = df.rename(columns={'resolution r': 'resolution r [m]'})
df = df.rename(columns={'RPE(m)': 'RPE [m]'})
df = df.rename(columns={'ATE(m)': 'ATE [m]'})
df = df.rename(columns={'bias-x(m)': 'bias-x [m]'})
df = df.rename(columns={'bias-y(m)': 'bias-y [m]'})
#df = df.rename(columns={'bias-y(m)': 'bias-y [m]'})
df = df.rename(columns={'bias-theta(deg)': 'bias-theta [deg]'})
cat_size_order = CategoricalDtype(
    [" P2L", " P2P", " P2D"], 
    ordered=True
)
df["cost type"] = df["cost type"].astype(cat_size_order)
df=df.sort_values("cost type")
#print(df)
##############################
```

```python
sns_plot = sns.catplot(x="submap keyframes", y="Transl. Error[%]", hue="cost type", kind="box", dodge=True, data=df, legend_out=False, palette=pal)
#sns_plot.legend(loc='upper right', ncol=1)
#lt.title('Transl. Error vs number of submap keyframes')
plt.ylim(1.1,3.05)
plt.grid()
plt.show()
sns_plot.savefig('./output/Transl_error_submap_keyframes.pdf', format='pdf',bbox_inches='tight')
sns_plot.savefig('./output/Transl_error_submap_keyframes.png', format='png', dpi=1500,bbox_inches='tight')
```

```python
sns_plot = sns.catplot(x="submap keyframes", y="Rot. Error[deg/100m]", hue="cost type", kind="box", dodge=True, data=df, legend_out=False, palette=pal)
#sns_plot.legend(loc='upper right', ncol=1)
#lt.title('Rotation. Error vs number of submap keyframes')
plt.ylim(0.34,0.68)
plt.grid()
plt.show()
sns_plot.savefig('./output/rot_error_submap_keyframes.pdf', format='pdf', box_inches='tight')
sns_plot.savefig('./output/rot_error_submap_keyframes.png', format='png', dpi=1500, box_inches='tight')
```

```python
sns_plot = sns.catplot(x="submap keyframes", y="RPE [m]", hue="cost type", kind="box", dodge=True, data=df, legend_out=False, palette=pal)
#sns_plot.legend(loc='upper right', ncol=1)
#lt.title('Relative Pose Error (RPE)[m] vs submap keyframes')
plt.grid()
plt.show()
sns_plot.savefig('./output/rpe_submap_keyframes.pdf', format='pdf',bbox_inches='tight')
sns_plot.savefig('./output/rpe_submap_keyframes.png', format='png', dpi=1500,bbox_inches='tight')
```

```python
sns_plot = sns.catplot(x="submap keyframes", y="bias-x [m]", hue="cost type", kind="box", dodge=True, data=df, legend_out=False, palette=pal)
#lt.title('longitudinal bias vs number of submap keyframes')
plt.axhline(0, color='red',linestyle='--')
plt.grid()
#plt.ylim(-0.028,0.02)


sns_plot.savefig('./output/bias_x_submap_keyframes.pdf', format='pdf',bbox_inches='tight')
sns_plot.savefig('./output/bias_x_submap_keyframes.png', format='png', dpi=1500,bbox_inches='tight')
```

```python

sns_plot = sns.catplot(x="submap keyframes", y="ATE [m]", hue="cost type", kind="box", dodge=True, data=df, legend_out=False, palette=pal)
plt.title('Absolute Trajectory Error (ATE) vs submap keyframes')
#sns_plot.legend(loc='upper right', ncol=1)
plt.ylim(10,80)
plt.grid()
plt.show()
sns_plot.savefig('./output/ATE_submap_keyframes.pdf', format='pdf',bbox_inches='tight')
sns_plot.savefig('./output/ATE_submap_keyframes.png', format='png', dpi=1500,bbox_inches='tight')
```
