```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from scipy.spatial.transform import Rotation as R
import pyautogui
from matplotlib import interactive
import matplotlib.patches as mpatches

interactive(True)
mutable_object = {} 
fig = plt.figure()
def LoadData(path):
    est = np.loadtxt(path)
    x=np.array(est[:,3])
    y=np.array(est[:,7])
    return (np.array([x,y]))
  
```

```python
############## VOLVO-CE ##################
#plt.rcParams['figure.figsize'] = [30, 30]
img = mpimg.imread('./volvo-CE/corrected_eskilstuna.jpg')

fig, ax = plt.subplots()
h, w, c = img.shape
plt.axis('off')

ax.imshow(img, extent=[0, w, 0, h])
#plt.rcParams['figure.dpi'] = 300


#print(path.shape)
path1=LoadData('./volvo-CE/cfear-1/est/01.txt')
path2=LoadData('./volvo-CE/cfear-2/est/01.txt')
path3=LoadData('./volvo-CE/cfear-3/est/01.txt')
r = R.from_euler('z',78, degrees=True)
Rot=r.as_matrix()[0:2,0:2]
trans=([[1255],[422]]) 
scale=2.41
#print(Rot)
#print(path)

path1 = np.matmul(Rot, path1*scale) + trans
path2 = np.matmul(Rot, path2*scale) + trans
path3 = np.matmul(Rot, path3*scale) + trans


plt.plot(path1[0, :],  path1[1, :],color="#F4D03F",linestyle="--", label="CFEAR-1",linewidth=1)
plt.plot(path2[0, :],  path2[1, :],color="#3498DB",linestyle="--", label="CFEAR-2",linewidth=1)
plt.plot(path3[0, :],  path3[1, :],color="#F7F9F9",linestyle="--", label="CFEAR-3",linewidth=1)
ax.legend( ['Cold', 'Medium', 'Hot', 'ground truth'])

CFEAR_1_patch = mpatches.Patch(color="#F4D03F", label='CFEAR-1')
CFEAR_2_patch = mpatches.Patch(color="#3498DB", label='CFEAR_2')
CFEAR_3_patch = mpatches.Patch(color="#F7F9F9", label='CFEAR-3')
CFEAR_4_patch = mpatches.Patch(color="#CB4335", label='Corrected trajectory')
plt.legend(handles=[CFEAR_1_patch, CFEAR_2_patch,CFEAR_3_patch, CFEAR_4_patch],loc=((1255-250)/w, (422+300)/h),fontsize=8)
plt.xlim(450, 1400)
plt.ylim(100, 900)


#plt.show()
plt.savefig('./output/volvo.jpg',format="jpg",dpi=200,bbox_inches='tight')
plt.savefig('./output/volvo.pdf',format="pdf",dpi=200,bbox_inches='tight') #,dpi=300
```
