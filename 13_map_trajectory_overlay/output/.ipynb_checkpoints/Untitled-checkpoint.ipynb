```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from scipy.spatial.transform import Rotation as R
import pyautogui
from matplotlib import interactive
import matplotlib.patches as mpatches

interactive(True)
mutable_object = {} 
fig = plt.figure()
def LoadData(path):
    est = np.loadtxt(path)
    x=np.array(est[:,3])
    y=np.array(est[:,7])
    return (np.array([x,y]))
def onclick(event):
    print('you pressed', event.key, event.xdata, event.ydata)
    X_coordinate = event.xdata
    Y_coordinate = event.ydata
    mutable_object['click'] = X_coordinate
    
plt.rcParams['figure.figsize'] = [30, 30]
img = mpimg.imread('./overlayed_eskilstuna.png')

fig, ax = plt.subplots()
h, w, c = img.shape

ax.imshow(img, extent=[0, w, 0, h])



#print(path.shape)
path1=LoadData('./volvo-CE/cfear-1/est/01.txt')
path2=LoadData('./volvo-CE/cfear-2/est/01.txt')
path3=LoadData('./volvo-CE/cfear-3/est/01.txt')
r = R.from_euler('z',78, degrees=True)
Rot=r.as_matrix()[0:2,0:2]
trans=([[1322],[439]]) 
scale=2.41
#print(Rot)
#print(path)

path1 = np.matmul(Rot, path1*scale) + trans
path2 = np.matmul(Rot, path2*scale) + trans
path3 = np.matmul(Rot, path3*scale) + trans


plt.plot(path1[0, :],  path1[1, :],color="#F4D03F", label="CFEAR-1",linewidth=3.0)
plt.plot(path2[0, :],  path2[1, :],color="#3498DB", label="CFEAR-2",linewidth=3.0)
plt.plot(path3[0, :],  path3[1, :],color="#F7F9F9", label="CFEAR-3",linewidth=3.0)
#ax.legend( ['Cold', 'Medium', 'Hot', 'ground truth'])

CFEAR_1_patch = mpatches.Patch(color="#F4D03F", label='CFEAR-1')
CFEAR_2_patch = mpatches.Patch(color="#3498DB", label='CFEAR_2')
CFEAR_3_patch = mpatches.Patch(color="#F7F9F9", label='CFEAR-3')
CFEAR_4_patch = mpatches.Patch(color="#CB4335", label='Corrected trajectory')

plt.legend(handles=[CFEAR_1_patch, CFEAR_2_patch,CFEAR_3_patch, CFEAR_4_patch],loc=((1322-300)/w, (439+200)/h),fontsize=20)

plt.show()
plt.savefig('./output/volvo.png')
plt.savefig('./output/volvo.pdf')
```
